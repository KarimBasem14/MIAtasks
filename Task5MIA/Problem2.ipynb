```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to plot
def f(x):
    # y = 1.4 sin(pi/2 x)
    x = (np.pi /2 ) * x
    sin = np.sin(x)
    y = 1.4 * sin
    return y
```

```python
# Create x values
# First part, 3bit encoder with 0.25 second time-sampling
"""
time = [0.25,0.5,0.75,1,1.25,1.5,1.75,2]
volt = [0.5357568053111257, 0.9899494936611666, 1.2934313455158013, 1.4, 1.2934313455158013, 0.9899494936611666, 0.5357568053111258, 0]
since the maximum voltage is 1.4 and we have 8 levels then each level can handle a volt range of 1.4/8 = 0.175
0.00 - 0.175 V : 000
0.175 - 0.35 V : 001
0.35 - 0.525 V : 010
0.525 - 0.7 V : 011
0.7 - 0.875 V : 100
0.875 - 1.05 V : 101
1.05 - 1.225 V : 110
1.225 - 1.4 V : 111
Now we can convert the samples list into binary code: [011,101,111,111,111,101,011,000]
final binary code = 011101111111111101011000
"""
x = np.linspace(0, 10, 1000)
y = f(x)

# Define the x-coordinate where you want to draw the vertical line
x_vert = 1

# Find the y-coordinate at the intersection
y_intersect = f(x_vert)
x_verts = [0.25,0.5,0.75,1,1.25,1.5,1.75,2]
volt = []
for n in x_verts:
    v = f(n)
    volt.append(v)
print(volt)
# Plot the function
plt.plot(x, y)

for x_vert in x_verts:
    # Find the y-coordinate at the intersection
    y_intersect = f(x_vert)
    
    # Draw the vertical line
    plt.axvline(x=x_vert, color='r', linestyle='--', label=f'x = {x_vert}')
    
    # Draw the horizontal line from the intersection point
    plt.axhline(y=y_intersect, color='b', linestyle='--', xmin=0, xmax=x_vert/2, label=f'y = sin({x_vert})')
    
    # Mark the point of intersection
    plt.plot(x_vert, y_intersect, 'go')
# Set the range for the y-axis and x-axis
plt.ylim(0, 1.5)
plt.xlim(0, 2)



# Display the plot
plt.show()
```
