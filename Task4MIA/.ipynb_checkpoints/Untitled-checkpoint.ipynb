```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# HEAT MAP NOT FINISHED AND TEMP VS TIME NOT FINISHED 
```

```python
df = pd.read_csv("./Weather_Dataset.csv") # Load the data
```

```python

# Filter the temp and humidity and visibility and cloud cover
df = df[(df['Temperature (C)'] > -100) & (df['Temperature (C)'] < 150) & (df['Humidity'] >= 0) & (df['Visibility (km)'] >= 0) & (df['Loud Cover'] >= 0)]
# Remove duplicates
df = df.drop_duplicates()

# Convert the 'Formatted Date' column to datetime format

# df['time'] = pd.to_datetime(df['Formatted Date'])
# print(df["time"])
# Plot the data
plt.figure(figsize=(12, 6))
sns.lineplot(x='Formatted Date', y='Temperature (C)', data=df)
plt.title('Temperature vs Time')
plt.xlabel('Time')
plt.ylabel('Temperature (°C)')
# plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a histogram of the temprature
temp_range = (min(df["Temperature (C)"]), max(df["Temperature (C)"]))
print(f"Min temp : {min(df["Temperature (C)"])}")
print(f"Max temp : {max(df["Temperature (C)"])}")
plt.hist(df["Temperature (C)"],range = temp_range)
plt.show()
```

```python
# Make a scatter plot of temperature vs humidity
temp = df["Temperature (C)"]
humidity = df["Humidity"]
print(min(df["Humidity"]))
print(max(df["Humidity"]))
plt.scatter(temp,humidity)
plt.show()
```

```python
# Create a corelation heatmap 
numeric_df = df.select_dtypes(include=[np.number]) # Take only the numeric values
corr_matrix = numeric_df.corr()
plt.figure(figsize=(10, 8))
# annot = True displays the correlation coeffectients numerically
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=5)
plt.title('Correlation Heatmap')
plt.show()
```

A correlation shows the relation between multiple things, are they connected to each other? If one thing happen how likely is it for another thing to happen? The correlation coeffecient ranges from -1 to 1 where 1 means perfect positive correlation (If one increases the other increases proportionally) , -1 means perfect negative correlation (if one increases the other decreases proportionally) and 0 means there is no relation at all.


Interpreting the graph

We are using a "coolwarm" color map so negative correlation is in blueish colors while positive correlation is in redish colors.
Deeper colors represents more correlation.
Now to understand the graph simply look at one variable on the Y-axis and draw a horizontal line in your head, then look at another variable on the X-axis and draw a vertical line in your head. Now see where they meet and that's the correlation between these 2 variables.
